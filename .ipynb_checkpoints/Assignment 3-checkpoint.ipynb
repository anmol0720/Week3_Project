# DATA NORMALISATION
    Normalising the datas enables us for proper comparision

  Method 1:Simple Feature scaling.
    Its value Ranges from 0 to 1

```python
#if we normalise population
max=df['population'].max()
df['population']=df['population']/max
df
```

Method 2:Min Max Method. Its value range from 0 to 1

```python
#Normalising Population
min=df['population'].min()
max=df['population'].max()
df['population']=(df['population']-min)/(max-min)
df
```

Method 3: Z-score calculation. Its value ranges from -3 to +3

```python
avg=df['population'].mean()
std=df['population'].std()
df['population']=(df['population']-avg)/std
df
```

# Analysis Of Variable(ANOVA)
    For Stastical Comparission of different variables
    Which Category effect the target the most
    Basically Correlation between target and other variable

It returns two values:
    
    1.F-test score-ANOVA assumes same mean for all groups and then compares the indivisual deviation. And the value we get from that is the F-test score.
                    Or, Ratio of variation between sample group means divided by variation within sample group.
                    
    2.P value- It tells us that the obtained value is stastically significant or not.

```python
from scipy import stats
df_anova=df[['suicides_no','suicides/100k pop']]

grouped_anova=df_anova.groupby(['suicides_no'])
anova_result=stats.f_oneway(grouped_anova.get_group(17)['suicides_no'],grouped_anova.get_group(481)['suicides_no'])
anova_result
```

If the ANOVA result gives a large F-test score and small P value then the correlation is strong

We Represent It By Using Regression Plot

```python
sns.regplot(x='suicides/100k pop',y='suicides_no',data=df)
plt.ylim(0,)
```

More Stepper The Regression Line. High is the Correlation

# Pearson Correlation
       - To measure the strength of correlation between two variables

It gives out two outputs:
    
    Correlation co-efficients:
         The value is as follows:
            Close to +1 : large +ve relation
            Close to -1 : large -ve relation
            Close to 0 : No Relation
    P-value(Probability Value):
         The valaue is as follows:
            <0.001 : Strong Certinity
            0.001 to 0.05 : Moderate Certinity
            0.05 to 0.1 : Weak Certinity
            >0.1 : No Certinity

```python
pearson_coff,p_value=stats.pearsonr(df['population'],df['suicides/100k pop'])
pearson_coff,p_value
```

The result is displayed as follows:
    
    (Correlation_coff,P-value)

Strong Correlation:
    
    Correlation Coefficient : Close to -1 to +1
    P-value : <0.001

HEAT-MAP for correlation
    
    Colour defines Strength

```python
df1=df[['suicides_no','population','suicides/100k pop']].corr()
print(df1)
plt.pcolor(df1,cmap='RdBu')
plt.colorbar()
```
